```
####################################
# CODIGOS TDAs -------------------------------------------------------------------------------------
####################################

# --------------------------------------------------------------------------------------------------
# TDA Auxilio --------------------------------------------------------------------------------------
# patente:Formato de Patente: 3 Letras 3 N√∫meros
# ZONAS: Sur=0 | Norte=1 | Este=2 | Oeste=3 | CABA=4 {ZonaAuxilio.name}
#   origen: Zona de partida
#   destino: Zona de llegada
# tipoDeAuxilio: Reparacion=0 | Remolque=1 {TipoAuxilio.name}
# estado: Espera=0 | Aprobado=1 {EstasoAux.name}

class Auxilio:
  def __init__(self,patente,origen,destino,tipoDeAuxilio,estado):
    self.patente = self.validaPatente(patente)
    self.origen = origen.name
    self.destino = destino.name
    self.tipoDeAuxilio = tipoDeAuxilio.name
    self.estado = estado.name
#-----------------------------------------------#
  def validaPatente(self,patente): # Valida la patente para almacenarla
    letras = patente[:3]
    numeros = patente[3:]
    if len(patente) !=6 or not letras.isalpha() or not numeros.isnumeric():
      raise Exception("PATENTE INVALIDA")
    return patente.upper() # Retorna la patente con la parte alfabetica en mayusculas 
#-----------------------------------------------#
  def __repr__(self):
    return "Patente Numero: {} [ Zona de Partida: {} | Zona de Llegada: {} | Tipo de Auxilio: {} | Estado: {} ]\n".format(self.patente,self.origen,self.destino,self.tipoDeAuxilio,self.estado)
#-----------------------------------------------#
  def getPatente(self):
    return self.patente
#-----------------------------------------------#
  def getOrigen(self):
    return self.origen
#-----------------------------------------------#
  def getDestino(self):
    return self.destino
#-----------------------------------------------#
  def getAuxTipo(self):
    return self.tipoDeAuxilio
#-----------------------------------------------#
  def getEstado(self):
    return self.estado
#-----------------------------------------------#
  def setTipoAux(self,nuevoTipo):
    self.tipoDeAuxilio = nuevoTipo.name
#-----------------------------------------------#
  def esReparacion(self):
    return self.getAuxTipo() == TipoAuxilio.Reparacion
#-----------------------------------------------#
  def esRemolque(self):
    return self.getAuxTipo() == TipoAuxilio.Remolque
#-----------------------------------------------#

# --------------------------------------------------------------------------------------------------
# --------------------------------------------------------------------------------------------------
# TDA OfcinaAtencion -------------------------------------------------------------------------------
# Definir un TDA que represente a una oficina donde se reciben las llamadas de
# los clientes, que maneja dos colas de auxilios, una de auxilios de "Remolque"
# y otra de auxilios de "Reparacion". Cada oficina se identifica con el interno
# en la central telef√≥nica (N√∫meros entre 1 y 999). Las oficinas tienen una
# cantidad cr√≠tica de auxilios de cada tipo (por defecto 50 pedidos).

class OficinaAtencion:
  def __init__(self,interno,auxiliosCriticos = 50):
    self.interno = self.validaEnteroEntre_Y_(interno, 1, 999, "INTERNO INEXISTENTE")
    self.colaDeRemolques = Queue()
    self.colaDeReparaciones = Queue()
    self.auxiliosCriticos = auxiliosCriticos  
#-----------------------------------------------#
  def __repr__(self):
    return "<"+str(self.interno)+">"
#-----------------------------------------------#
  def getRemolques(self):
    return self.colaDeRemolques
#-----------------------------------------------#
  def getReparaciones(self):
    return self.colaDeReparaciones
#-----------------------------------------------#
  def validaEnteroEntre_Y_(self, valor, minimo, maximo, mensaje= "VALOR INESPERADO"):
    if valor < minimo or valor >= maximo :
      raise Exception(mensaje)
    return valor
#-----------------------------------------------#
  def recibirAuxilio(self,auxilio):
    if auxilio.esReparacion():
      self.colaDeReparaciones.enqueue(auxilio)
      if self.colaDeReparaciones.long()>= self.auxiliosCriticos:
        print("ADVERTENCIA: SE HA SUPERADO LA CANTIDAD CRITICA DE REPARACIONES")
    else:
      self.colaDeRemolques.enqueue(auxilio)
      if self.colaDeRemolques.long()>= self.auxiliosCriticos:
        print("ADVERTENCIA: SE HA SUPERADO LA CANTIDAD CRITICA DE REMOLQUES")
#-----------------------------------------------#
  def primerAuxilioAEnviar(self):
    if not self.colaDeRemolques.isEmpty():
      return self.colaDeRemolques.top()
    elif not self.colaDeReparaciones.isEmpty():
      return self.colaDeReparaciones.top()
#-----------------------------------------------#
  def enviarAuxilio(self,zonaDeGrua):
    pedido = None
    colaAux = Queue()
    while not self.colaDeReparaciones.isEmpty() and zonaDeGrua.name != self.primerAuxilioAEnviar().getOrigen():
      if not self.colaDeRemolques.isEmpty():
        colaAux.enqueue(self.colaDeRemolques.dequeue())
      else:
        colaAux.enqueue(self.colaDeReparaciones.dequeue())
    
    if not self.colaDeRemolques.isEmpty():
      pedido = self.colaDeRemolques.dequeue()
    else:
      pedido = self.colaDeReparaciones.dequeue()

    while not colaAux.isEmpty():
      elem = colaAux.dequeue()
      if elem.esReparacion():
        self.colaDeReparaciones.enqueue(elem)
      else:
        self.colaDeRemolques.enqueue(elem)
    return pedido
#-----------------------------------------------#
  def auxiliosPorTipo(self):
    return self.colaDeRemolques.long(), self.colaDeReparaciones.long()
#-----------------------------------------------#
  def cantidadTotalAuxilios(self):
    return self.colaDeRemolques.long() + self.colaDeReparaciones.long()
#-----------------------------------------------#
  def esCritica(self):
    return self.colaDeRemolques.long()>= self.auxiliosCriticos or self.colaDeReparaciones.long()>= self.auxiliosCriticos
#-----------------------------------------------#
  def auxiliosEnEspera(self):
    cuentaEspera = 0
    colaAux = Queue()
    while not self.colaDeReparaciones.isEmpty():
      if self.primerAuxilioAEnviar().getEstado() == EstadoAuxilio(0).name:
        cuentaEspera += 1

      if not self.colaDeRemolques.isEmpty():
        colaAux.enqueue(self.colaDeRemolques.dequeue())
      else:
        colaAux.enqueue(self.colaDeReparaciones.dequeue())
    
    while not colaAux.isEmpty():
      elem = colaAux.dequeue()
      if elem.esReparacion():
        self.colaDeReparaciones.enqueue(elem)
      else:
        self.colaDeRemolques.enqueue(elem)

    return cuentaEspera
#-----------------------------------------------#
  def buscarAuxilio(self,nroPatente):
    encontrado = self.buscarPatenteEnCola(nroPatente,self.colaDeRemolques)
    if encontrado == None:
      encontrado = self.buscarPatenteEnCola(nroPatente,self.colaDeReparaciones)
    return encontrado

  def buscarPatenteEnCola(self,nroPatente,unaCola):
    encontrado = None
    patenteMayuscula = nroPatente.upper()
    colaAux = unaCola.clone()
    while not colaAux.isEmpty():
      elem = colaAux.dequeue()
      if patenteMayuscula == elem.getPatente():
        encontrado = elem
    return encontrado
#-----------------------------------------------#
  def eliminarAuxilio(self,nroPatente):
    colaAux = Queue()
    buscado = self.buscarAuxilio(nroPatente)
    while buscado != None and buscado != self.primerAuxilioAEnviar():
  
      if not self.colaDeRemolques.isEmpty():
        colaAux.enqueue(self.colaDeRemolques.dequeue())
      else:
        colaAux.enqueue(self.colaDeReparaciones.dequeue())
    
    if not self.colaDeRemolques.isEmpty():
      self.colaDeRemolques.dequeue()
    else:
      self.colaDeReparaciones.dequeue()
    
    while not colaAux.isEmpty():
      elem = colaAux.dequeue()
      if elem.esReparacion():
        self.colaDeReparaciones.enqueue(elem)
      else:
        self.colaDeRemolques.enqueue(elem)
#-----------------------------------------------#
  def cambiaDeTipo(self,nroPatente):
    buscado = self.buscarAuxilio(nroPatente)
    if buscado != None:
      colaAux = Queue()
      while buscado != self.primerAuxilioAEnviar():
    
        if not self.colaDeRemolques.isEmpty():
          colaAux.enqueue(self.colaDeRemolques.dequeue())
        else:
          colaAux.enqueue(self.colaDeReparaciones.dequeue())
      
      if not self.colaDeRemolques.isEmpty():
        elem = self.colaDeRemolques.dequeue()
        elem.setTipoAux(TipoAuxilio.Reparacion)
      else:
        elem = self.colaDeReparaciones.dequeue()
        elem.setTipoAux(TipoAuxilio.Remolque)

      self.recibirAuxilio(elem)

      while not colaAux.isEmpty():
        elem = colaAux.dequeue()
        if elem.esReparacion():
          self.colaDeReparaciones.enqueue(elem)
        else:
          self.colaDeRemolques.enqueue(elem)
#-----------------------------------------------#


# --------------------------------------------------------------------------------------------------
# --------------------------------------------------------------------------------------------------
# TDA EdificioEmpresa ------------------------------------------------------------------------------
# En el edificio de la empresa tenemos M pisos con N habit√°culos cada uno (un
# habit√°culo es un lugar f√≠sico). En cada habit√°culo se ubica una oficina de
# atenci√≥n telef√≥nica con su n√∫mero de interno. Modelar con un TDA el Edificio
# de la Empresa, teniendo en cuenta que puede haber habit√°culos vacios, cuando
# un habit√°culo esta vac√≠o, se representa como None en la estructura.

class EdificioEmpresa:
  def __init__(self, piso = 0, habitaculo = 0):
    self.edificio = np.empty((piso,habitaculo),OficinaAtencion)
#-----------------------------------------------#
  def __repr__(self):
    return "EDIFICIO:\n" + str(self.edificio)
#-----------------------------------------------#
  def establecerOficina(self,numeroPiso, numeroHabitaculo, oficinaAtencion):
    self.edificio[numeroPiso,numeroHabitaculo] = oficinaAtencion
#-----------------------------------------------#
  def getOficina(self,numeroPiso,numeroHabitaculo):
    return self.edificio[numeroPiso][numeroHabitaculo]
#-----------------------------------------------#
  def cantidadDeOficinasCriticas(self,piso): # NO ES RECURSIVA
    contadorCriticos=0
    for i in range (len (self.edificio[piso])):
      if self.edificio[piso][i] != None and self.edificio[piso][i].esCritica():
          contadorCriticos = contadorCriticos + 1
    return contadorCriticos

  #####################################################
 ##############  N O   R E S U E L T O  ############## Presenta problemas con el caso base
#####################################################  No se logro determinar
  def NO_CORRE_cantidadDeOficinasCriticas(self,piso):
    totalCriticas = None
    if len(self.edificio[piso]) <= 0:
      totalCriticas = 0
    elif len(self.edificio[piso]) == 1:
      oficina = self.edificio[piso][0]
      if oficina != None and oficina.esCritica():
        totalCriticas = 1
      else:
        totalCriticas = 0
    else:
      habitaculo = len(self.edificio[piso])-1
      oficina = self.edificio[piso][habitaculo]
      if oficina != None and oficina.esCritica():
        totalCriticas = self.cantidadDeOficinasCriticas(piso) + 1
    return totalCriticas
#-----------------------------------------------#
  def oficinaMenosRecargada(self):
    nroPiso = None
    nroHabitaculo = None
    minimo = 0
    for piso in range(len(self.edificio)):
      for habitaculo in range(len(self.edificio[piso])):
        oficina = self.edificio[piso][habitaculo]   
        if oficina != None:
          remolque,reparacion = oficina.auxiliosPorTipo()     
          if oficina != None and minimo > remolque:
            minimo = remolque
            nroPiso = piso
            nroHabitaculo = habitaculo
    return nroPiso,nroHabitaculo
#-----------------------------------------------#
  def buscaOficina(self,nroInterno):
    nroPiso = None
    nroHabitaculo = None
    for piso in range(len(self.edificio)):
      for habitaculo in range(len(self.edificio[piso])):
        oficina = self.edificio[piso][habitaculo]
        if (oficina is not None) and (nroInterno == oficina.interno):
          nroPiso = piso
          nroHabitaculo = habitaculo
    return nroPiso,nroHabitaculo
#-----------------------------------------------#
  def centralTelefonica(self,pilaDeAuxilios):
    nroPiso,nroHabitaculo = self.oficinaMenosRecargada()
    while not pilaDeAuxilios.isEmpty():
      unAuxilio = pilaDeAuxilios.pop()
      self.edificio[nroPiso][nroHabitaculo].recibirAuxilio(unAuxilio)
#-----------------------------------------------#
  def moverAuxilio(self,nroPatente,internoOficinaOrigen,internoOficinaDestino):
    nroPisoOrigen,nroHabitaculoOrigen = self.buscaOficina(internoOficinaOrigen)
                # Guarda en la Tupla las coordenadas de la oficina origen en la matriz del edificio
    oficina = self.edificio[nroPisoOrigen][nroHabitaculoOrigen]
                # Guarda en "oficina" la informacion de esta segun sus coordenadas
    unAuxilio = oficina.buscarAuxilio(nroPatente)
                # Busca el auxilio por patente en la oficina y lo guarda en la variable unAuxilio
    self.edificio[nroPisoOrigen][nroHabitaculoOrigen].eliminarAuxilio(nroPatente)
                # Elimina el auxilio de la cola de auxilios de la oficina de origen
    nroPisoDestino,nroHabitaculoDestino = self.buscaOficina(internoOficinaDestino)
                # Identifica las coordenadas de la oficina destino en la matriz del edificio
    self.edificio[nroPisoDestino][nroHabitaculoDestino].recibirAuxilio(unAuxilio)
                # Recibe el auxilio en la cola de la oficina destino
#-----------------------------------------------#

# --------------------------------------------------------------------------------------------------
# --------------------------------------------------------------------------------------------------


####################################
# CODIGOS Enums ------------------------------------------------------------------------------------
####################################

#---------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------
import numpy as np
from enum import Enum
#-----------------------------------------------#
class TipoAuxilio(int,Enum):
  Remolque = 0
  Reparacion = 1
#-----------------------------------------------#
class ZonaAuxilio(int,Enum):
  Sur = 0
  Norte = 1
  Este = 2
  Oeste = 3
  CABA = 4
#-----------------------------------------------#
class EstadoAuxilio(int,Enum):
  Espera = 0
  Aprobado = 1
#-----------------------------------------------#

# --------------------------------------------------------------------------------------------------
# --------------------------------------------------------------------------------------------------


####################################
# CODIGOS AUXILIARES / REUTILIZADOS ----------------------------------------------------------------
####################################

#---------------------------------------------------------------------------------------------------
# --------------------------------------------------------------------------------------------------
# TDA Stack (Pila) ---------------------------------------------------------------------------------
# 
class Stack:
  def __init__(self):
    self.pila = []
#-----------------------------------------------#
  def vaciar(self): # Vac√≠a la pila
    self.pila.clear()
#-----------------------------------------------#
  def push(self,elemento): # Apila un elemento
    self.pila.append(elemento)
#-----------------------------------------------#
  def pop(self): # Desapila un elemento
    if not self.isEmpty():
      return self.pila.pop()
#-----------------------------------------------#
  def top(self): # Obtener primer elemento
    if not self.isEmpty():
      return self.pila[-1]
#-----------------------------------------------#
  def clone(self): # Clonar una pila
    unaPila = Stack()
    for elemento in self.pila:
      unaPila.push(elemento)
    return unaPila
#-----------------------------------------------#
  def long(self): # Tama√±o de una pila
    return len(self.pila)
#-----------------------------------------------#
  def isEmpty(self): # Retorna booleano si la pila est√° vacia
    return self.long() == 0
#-----------------------------------------------#
  def __repr__(self): # Representacion de una pila
    return str(self.pila)
#-----------------------------------------------#


#---------------------------------------------------------------------------------------------------
# --------------------------------------------------------------------------------------------------
# TDA Queue (Cola) ---------------------------------------------------------------------------------
# 
class Queue:
  def __init__(self):
    self.cola = []
#-----------------------------------------------#
  def vaciar(self): # Vacia cola
    self.cola.clear()
#-----------------------------------------------#
  def enqueue(self,elemento): # Encola un elemento
    self.cola.insert(0,elemento)
#-----------------------------------------------#
  def dequeue(self): # Desencola un elemento
    if not self.isEmpty():
      return self.cola.pop()
#-----------------------------------------------#
  def top(self): # Obtenet primer elemento
    if not self.isEmpty():
      return self.cola[-1]
#-----------------------------------------------#
  def clone(self): # Clonar una cola
    unaCola = Queue()
    for elemento in self.cola:
      unaCola.enqueue(elemento)
    return unaCola
#-----------------------------------------------#
  def long(self): # Obtener tama√±o de una cola
    return len(self.cola)
#-----------------------------------------------#
  def isEmpty(self): # Retorna booleano si la cola est√° vacia
    return self.long() == 0
#-----------------------------------------------#
  def __repr__(self): # Representacion de una cola
    return str(self.cola)
#-----------------------------------------------#
```
